This line checks which users have rated more than 200 books

Extracts the User-IDs of users who have rated more than 200 books.

```python
wellread_users = x[x].index
```

This line filters the ratings DataFrame to include only those entries made by well-read users (who rated more than 200 books).

```python
filtered_rating = df_merge[df_merge['User-ID'].isin(wellread_users)]
filtered_rating
```

This line checks which books have been rated at least 40 times by well-read users 

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=40
y
```

```python
famous_books = y[y].index
famous_books
```

This line filters the ratings to include only those for books that are considered famous (rated at least 40 times by well-read users)

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

Make a pivot table

```python
pvt = final_ratings.pivot_table(index="Book-Title",columns="User-ID",values='Book-Rating')
pvt
```

```python
pvt.fillna(0,inplace=True)
```

Lets computes the cosine similarity between all book vectors based on user ratings from the pivot table.

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_scores = cosine_similarity(pvt)
```

```python
similarity_scores.shape
```

```python
def recommend_book(name):
    """This function takes a book title as input and returns a list of the top 5
    most similar books based on cosine similarity of user ratings."""
    index = np.where(pvt.index==name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pvt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        data.append(item)

    return data
```

```python
recommend_book("Women Who Run with the Wolves")
```
